```python
#Assinging values in X & Y for feature selection
X = df[['R&D Spend', 'Administration', 'Marketing Spend','State']]
y = df['Profit']
```

```python
X
```

```python
y
```

```python
#state contains categorical data, it should be encoded before building the model by using the dummy variable trap
#So to avoid the dummy variable trap we have to drop one dummy variable out of 3 while building the model
X=pd.get_dummies(X, drop_first=True)
#Changing bollean to integer
X = X.replace({True: 1, False: 0})
```

```python
X
```

```python
#Splitting testdata into X_train,X_test,y_train,y_test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=2)#perfroms the split
print(X_train)
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#NOW MODEL DEVELOPMENT
#Model Development the process of finding the exemplary model architecture, creating and training, and evaluating the results
#As per EDA done earlier, our data forms a Gaussian distribution, which is linearly separable and also good fit for linear regression model


from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

```

```python
#Predicting Results as our model is ready, let’s now check the prediction.
y_pred = lr.predict(X_test)
print(y_pred)
```

```python
#here we can see difference between predicted and actual y can be seen
df=pd.DataFrame(data={'Predicted value':y_pred,'Actual Value':y_test})
df
```

```python
#Evaluating Performance Based On Model Score by a process model evaluation

#testing data model
testing_data_model_score = lr.score(X_test, y_test)
print("Model Score/Performance on Testing data",testing_data_model_score)

#training data model 

training_data_model_score = lr.score(X_train, y_train)
print("Model Score/Performance on Training data",training_data_model_score)

#We can see that our model is 97% accurate on test data and 93% accurate with train data
```

```python
#Evaluating Performance Based On Metric such as R², MSE, RMSE, and MAE

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
 
mae = mean_absolute_error(y_pred,y_test)
print("Mean Absolute Error is :" ,mae)
 
mse = mean_squared_error(y_pred, y_test)
print("Mean Squarred Error is :" ,mse*100)

rmse = np.sqrt(mean_squared_error(y_pred, y_test))
print("Root Mean Squarred Error is :",rmse*100)
 
r2Score = r2_score(y_pred, y_test)
print("R2 score of model is :" ,r2Score*100)
#multiplying by 100 insures we get a result between -100 to +100, instead of -1 to +1
#r2 value is 0.97 pretty close to 1 which is a good indication of good model performance
```
