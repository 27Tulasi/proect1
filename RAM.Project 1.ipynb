```python
#replacing integer with string 
dfbudget_domesticgross = df4['domestic_gross'].str.replace('[\$,]', '', regex=True).astype('int64')
dfbudget_worldgross = df4['worldwide_gross'].str.replace('[\$,]', '', regex=True).astype('int64')
```

```python
# Caculate total gross
df4['total_gross'] = dfbudget_domesticgross + dfbudget_worldgross
```

```python
df4['total_gross']
```

```python
# now Extracting month from the release date by using str.split function
df4['month'] = df4['release_date'].str.split(' ', expand=True)[0].astype(str)
df4['month'] = df4['month'].str.strip()
```

```python
#if you run now it will show all the changes made to the actual file
df4
```

```python
# Calculate the average of total gross by each month with mean
df_moviebestmonth = df4.groupby('month')['total_gross'].mean()

# Make the data more understandable
df_moviebestmonth = pd.DataFrame(df_moviebestmonth)
df_moviebestmonth = df_moviebestmonth.reset_index()
df_moviebestmonth.columns = ['month', 'total_gross']
df_moviebestmonth
```

```python
# Replace the months in number format for ranking using replace function

df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Jan': '1'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Feb': '2'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Mar': '3'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Apr': '4'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'May': '5'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Jun': '6'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Jul': '7'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Aug': '8'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Sep': '9'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Oct': '10'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Nov': '11'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].replace({'Dec': '12'})
df_moviebestmonth['month'] = df_moviebestmonth['month'].astype(int)
df_moviebestmonth = df_moviebestmonth.sort_values('month')
df_moviebestmonth
```

```python
#now create the bar chart to compare the gross for each month
df_moviebestmonth.plot.bar(x='month', y='total_gross', title='Average gross for each month')
plt.xticks(rotation=0)
plt.show()
```

```python
#combining the data in different tables to get the Return Of Investment
df_genre_ROI = df4.merge(df2[['primary_title', 'genres']], 
                           left_on='movie', 
                           right_on='primary_title', 
                           how='inner')
df_genre_ROI
```

```python
#getting the insights of production budget 
df_genre_ROI['production_budget'] = df_genre_ROI['production_budget'].str.replace('[\$,]', '', regex=True).astype('int64')
df_genre_ROI['ROI'] = (df_genre_ROI['total_gross'] - df_genre_ROI['production_budget'])/df_genre_ROI['production_budget']
df_genre_ROI
```

```python
#getting the Average
df_genre_ROI_grouped = df_genre_ROI.groupby('genres')[['ROI', 'production_budget']].mean()
df_genre_ROI_grouped
```

```python
# to get the data more understandable and ROI in % format
df_genre_ROI_grouped = pd.DataFrame(df_genre_ROI_grouped)
df_genre_ROI_grouped = df_genre_ROI_grouped.reset_index()
df_genre_ROI_grouped.columns = ['Genres', 'ROI(%)', 'Budget($)']
df_sorted = df_genre_ROI_grouped.sort_values('ROI(%)', ascending=False)
df_sorted.head(20)
```

```python
#according to one survey movies with >27-29% of ROI considerd as success
df_sorted['ROI(%)'] = df_sorted['ROI(%)'].astype(float)
df_sorted_goodROI = df_sorted[df_sorted['ROI(%)'] > 29]
df_sorted_goodROI
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

df_sorted_goodROI['Budget($)'] = df_sorted_goodROI['Budget($)'] / 1e6 

plt.figure(figsize=(18, 8))  
sns.scatterplot(data=df_sorted_goodROI, x='Budget($)', y='ROI(%)', hue='Genres', palette='tab10', s=150)

plt.xlabel('Budget($Million))') 
plt.ylabel('ROI(%)')  

plt.xticks(ticks=np.linspace(df_sorted_goodROI['Budget($)'].min(), df_sorted_goodROI['Budget($)'].max(), 10))
plt.yticks(ticks=np.linspace(df_sorted_goodROI['ROI(%)'].min(), df_sorted_goodROI['ROI(%)'].max(), 10))

plt.legend(bbox_to_anchor=(1.05, 1), loc=2)  
plt.show()
```
